```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/car_price_dataset.csv")
df
```

```python
#Q1.What is the average price of all cars in the dataset?
##ANSWER:-
df['Price'].mean()
print(f"this is the average price of all cars in dataset is ${round(df['Price'].mean(),2)}")
```

```python
#Q2.Which car brand has the highest average price?
##ANSWER:-
brand=df[df['Price']==df['Price'].max()]["Brand"][1100]
print(f"this is the car brand with highest average price is {brand}")
```

```python
#Q3.What is the most common fuel type in the dataset?
##ANSWER:-
fuel=df['Fuel_Type'].value_counts().idxmax()
print(f"this is the most common fuel type in the dataset is {fuel}")
```

```python
#Q4.How many cars have a mileage greater than 200,000?
##ANSWER:-
count=df[df["Mileage"]>200000].shape[0]
print(f"Number of cars greater average than 200000 is: {count}")
```

```python
#Q5.What is the distribution of cars by transmission type?
##ANSWER:-
transmission_type=df['Transmission'].value_counts()
print(f"distribution of cars by transmission type is:\n{transmission_type}")



```

```python
#Q6.Which car model has the highest price?
##ANSWER:-
car=df.loc[df["Price"].idxmax(),"Model"]
price=df["Price"].max()
print(f"this is the car model with highest price is {car}")
print(f"this is the price of car is ${price}")
```

```python
#Q7.What is the average price of cars by fuel type?
##ANSWER:-
average_price=df.groupby("Fuel_Type")["Price"].mean()
print(f"average price of cars by fuel type is:\n{average_price}")
```

```python
#Q8.How many cars are from the year 2020 or later?
##ANSWER:-
cars=df[df["Year"]>=2020].shape[0]
print(f"this is the number of cars are from the year 2020 or later is {cars}")
```

```python
#Q9.What is the average engine size for each brand?
##ANSWER:-
average_engine = [df.groupby("Brand")["Engine_Size"].mean()]
print(f"average engine size for each brand is:\n{average_engine}")

```

```python
#Q10.Which car has the lowest mileage?
##ANSWER:-
lowest_mileage=df.loc[df["Mileage"].idxmin(),"Model"]
mileage=df["Mileage"].min()
print(f"this is the car with lowest mileage is {lowest_mileage}")
print(f"this is the mileage of car is {mileage}")
```

```python
#Q11.What is the correlation between mileage and price?
##ANSWER:-
correlation=df["Mileage"].corr(df["Price"])
print(f"this is the correlation between mileage and price is {correlation}")
```

```python
#Q13.What is the average price of cars by year?
##ANSWER:-
average_price=df.groupby("Year")["Price"].mean()
print(f"average price of cars by year is:\n{average_price}")
```

```python
#Q15.What is the average mileage for each fuel type?
##ANSWER:-
average_mileage=df.groupby("Fuel_Type")["Mileage"].mean()
print(f"average mileage for each fuel type is:\n{average_mileage}")
```

```python
#Q16.How many cars are from the brand 'Toyota'?
#ANSWER:-
cars=df[df["Brand"]=="Toyota"].shape[0]
print(f"this is the number of cars are from the brand 'Toyota' is {cars}")
```

```python
#Q17.What is the average price of cars with automatic transmission?
##ANSWER:-
average_price=df[df["Transmission"]=="Automatic"]["Price"].mean()
print(f"average price of cars with automatic transmission is ${round(average_price,2)}")
```
